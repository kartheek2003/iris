In the **Breast Cancer Wisconsin (Diagnostic) Dataset**, the target variable **`Diagnosis`** has two possible values:

- **`B`**: **Benign** (non-cancerous)
- **`M`**: **Malignant** (cancerous)

These labels indicate whether a tumor is benign (harmless) or malignant (harmful), which is what the model aims to predict. The goal is to classify a tumor as either benign or malignant based on 30 numeric features derived from a digitized image of a fine-needle aspirate (FNA) of a breast mass.

### Example:
- **Benign (B)** tumors tend to grow slowly and are less likely to spread.
- **Malignant (M)** tumors can grow rapidly and spread to other parts of the body.

```python
df['diagnosis'] = df['diagnosis'].replace({'M':1,'B':0})
```

In the **Breast Cancer Wisconsin (Diagnostic) Dataset**, the features are derived from digitized images of fine-needle aspirates (FNA) of breast tumors. These features are calculated for each cell nucleus and grouped into three categories based on statistical properties:

### 1. **Mean (`mean`)**
   - Represents the **average** value of the feature across all nuclei in the image.
   - Example: `radius_mean`, `texture_mean`, `area_mean`.

### 2. **Standard Error (`se`)**
   - Represents the **variation** or **uncertainty** in the measurement of the feature.
   - Standard error gives insight into how consistent the feature measurement is.
   - Example: `radius_se`, `texture_se`, `area_se`.

### 3. **Worst (`worst`)**
   - Represents the **largest** (or most extreme) value of the feature across all nuclei in the image.
   - It's usually derived from the top 10% largest values of the feature.
   - Example: `radius_worst`, `texture_worst`, `area_worst`.

### Why These Categories are Important:
- **Mean** features provide an overall average measurement, giving a broad sense of the tumor characteristics.
- **Standard Error** features indicate variability, helping in assessing the consistency of the tumor's features.
- **Worst** features highlight extreme cases, which can be critical in identifying malignant tumors with highly irregular shapes or large cell sizes.

### Example of Features:
| Feature Name      | Description                                        |
|-------------------|----------------------------------------------------|
| `radius_mean`      | Average radius of nuclei in the tumor              |
| `texture_se`       | Variation in texture                               |
| `perimeter_worst`  | Largest perimeter value across nuclei              |

By including these three types of measurements, the dataset captures a comprehensive view of the tumor's characteristics, improving the model's ability to differentiate between benign and malignant tumors.

The dataset is slightly imbalanced, with more benign cases than malignant ones. While this imbalance isn't severe, it could still influence model performance, particularly for metrics like accuracy.


Resampling:

Oversample the minority class (Malignant) using techniques like SMOTE (Synthetic Minority Oversampling Technique).

Undersample the majority class (Benign) to balance the dataset.


```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
```

```python
x_res,y_res = smote.fit_resample(x,y)
```

Why Scale After SMOTE?

Preserves Data Consistency: Synthetic samples from SMOTE will have the same range as the original data but scaling afterward ensures all features are on the same scale.

Improves Model Performance: Scaling is essential for algorithms sensitive to distances and gradients, like SVM, Logistic Regression, and KNN.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x_res)
```

 ### Hyperparameter Tuning

```python
import warnings
warnings.filterwarnings("ignore")


import numpy as np
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# Define parameter grids

# Logistic Regression
param_grid_logistic = {
    'C': np.logspace(-4, 4, 10),
    'penalty': ['l1', 'l2', 'elasticnet'],
    'solver': ['liblinear', 'saga'],
}

# Support Vector Machine
param_grid_svc = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly', 'sigmoid'],
    'gamma': ['scale', 'auto'],
}

# K-Nearest Neighbors
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski'],
}

# Random Forest
param_grid_rf = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False],
}

# XGBoost
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
}

# Model definitions
models = {
    'LogisticRegression': (LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42), param_grid_logistic),
    'SVC': (SVC(probability=True, class_weight='balanced', random_state=42), param_grid_svc),
    'KNN': (KNeighborsClassifier(), param_grid_knn),
    'RandomForest': (RandomForestClassifier(class_weight='balanced', random_state=42), param_grid_rf),
    'XGBoost': (XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42), param_grid_xgb),
}

# Run RandomizedSearchCV for each model
for model_name, (model, param_grid) in models.items():
    randomized_search = RandomizedSearchCV(
        estimator=model,
        param_distributions=param_grid,
        n_iter=50,  # Number of parameter settings sampled
        scoring='roc_auc',
        cv=5,
        verbose=1,
        random_state=42,
        n_jobs=-1
    )
    
    randomized_search.fit(x_train_pca, y_train)
    
    print(f"Best parameters for {model_name}:")
    print(randomized_search.best_params_)
    print(f"Best ROC-AUC score for {model_name}: {randomized_search.best_score_}\n")
```

Why Use an Ensemble?

Better Performance: Combines the strengths of multiple models, reducing individual model weaknesses.

Robustness: Handles variance and bias better than individual models.

Versatility: Balances models that perform well in different areas (e.g., SVC handles margins, Random Forest handles trees).


```python
from sklearn.ensemble import VotingClassifier
```

```python
# Models with tuned parameters
logistic = LogisticRegression(solver='saga', penalty='l1', C=0.359, class_weight='balanced', random_state=42)
svc = SVC(kernel='rbf', gamma='auto', C=10, probability=True, random_state=42)
knn = KNeighborsClassifier(weights='distance', n_neighbors=7, metric='manhattan')
random_forest = RandomForestClassifier(n_estimators=200, max_depth=20, min_samples_split=2, min_samples_leaf=1, bootstrap=False, random_state=42)
xgboost = XGBClassifier(subsample=0.7, n_estimators=200, max_depth=3, learning_rate=0.1, colsample_bytree=0.8, random_state=42)
```

```python
# Ensemble using soft voting (probability-based voting)
voting_ensemble = VotingClassifier(
    estimators=[
        ('logistic', logistic),
        ('svc', svc),
        ('knn', knn),
        ('random_forest', random_forest),
        ('xgboost', xgboost)
    ],
    voting='soft'  # 'soft' uses predicted probabilities; 'hard' uses predicted labels
)
```

```python
# Train the ensemble model
voting_ensemble.fit(x_train_pca, y_train)
```

```python
# Predictions
y_pred = voting_ensemble.predict(x_test_pca)
y_proba = voting_ensemble.predict_proba(x_test_pca)[:, 1]
```

```python
from sklearn.metrics  import classification_report , confusion_matrix , roc_auc_score
```

```python
# Evaluation
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```
